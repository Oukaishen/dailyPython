```python
import numpy as np
```

```python
MarkovA = np.array([[0.9,0.075,0.025],[0.15,0.8,0.05],[0.25,0.25,0.5]], dtype=float).T
MarkovA
```

```python
v0 = np.array([[0.3,0.4,0.3]], dtype=float).T
v0
```

```python
for i in range(100):
    v0 = MarkovA.dot(v0)
    print("Current round:" , i+1)
    print(v0)
```

```python
w, v = np.linalg.eig(MarkovA)
```

```python
print(w)
print(v)
```
